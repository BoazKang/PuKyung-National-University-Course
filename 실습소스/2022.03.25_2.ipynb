```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
plt.ylim(-2,2)
plt.xlim(-1, 11)
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')
ax.legend(loc='upper left')
```
