```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_boston
X = load_boston().data
y = load_boston().target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
svr_tr_model = SVR()
svr_model = SVR()
```

```python
svr_tr_model.fit(X_train, y_train)
svr_model.fit(X, y)
```

```python
svr_total_score = mean_squared_error(svr_model.predict(X), y)
svr_tr_score = mean_squared_error(svr_tr_model.predict(X_test), y_test)
print(svr_total_score,'\n',svr_tr_score)
```

```python
svr_model.get_params()
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(X,y)
rf_tr_model = RandomForestRegressor()
rf_tr_model.fit(X_train,y_train)
```

```python
mean_squared_error(rf_model.predict(X),y)
```

```python
mean_squared_error(rf_tr_model.predict(X_test),y_test)
```

```python
lm = LinearRegression()
lm.fit(X,y)
lm_tr = LinearRegression()
lm_tr.fit(X_train, y_train)
```

```python
mean_squared_error(lm.predict(X),y)
```

```python
mean_squared_error(lm_tr.predict(X_test),y_test)
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.datasets import load_digits
digits = load_digits()

X = digits.data
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
dt_model = DecisionTreeClassifier()
dt_tr_model = DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
dt_tr_model.fit(X_test, y_test)
```

```python
dt_model.score(X_test, y_test)
```

```python
for i in np.arange(5):
    print(f'실제값은 : {digits.target[i]}')
    print(f'예측값은 : {dt_model.predict(digits.data[[i]])}')
    plt.show()
    plt.imshow(digits.images[i])
```
