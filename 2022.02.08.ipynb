```python
def welcome(name):
    print(f"총 {len(name)}명의 학생이 존재합니다")
    while True:
        choice1 = random.choice(name)
        print(f'{choice1} 님')
        name.remove(choice1)
        if len(name) == 2:
            break
    print("이 선택되었습니다 축하합니다")
```

```python
def welcome(x):
    print("총 {}명의 학생이 존재합니다".format(len(x)))
    while True:
        choice1 = random.choice(x)
        print('{} 님'.format(choice1))
        choice2 = random.choice(x)
        if choice1 != choice2:
            print('{} 님'.format(choice2))
            break
        else:
            tmp_x = [i for i in x if i != choice1]
            choice2 = random.choice(tmp_x)
            print("{} 님".format(choice2))
            break
    print("이 선택되었습니다 축하합니다")
```

```python
def fib(n):
    f_1 = 1
    f_2 = 1
    tmp = [f_1, f_2]
    for i in range(n):
        tmp.append((tmp[i]+tmp[i+1]))
    return tmp[-3]
```

```python
fib(10)
```

```python
def fibo(n):
    if (n == 1) or (n == 2):
        return 1
    return fibo(n-1) + fibo(n-2)
```

```python
fibo(7)
```

```python
counter = 0
def fibo(n):
    global counter
    counter += 1
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
```

```python
def flatten(x):
    result = []
    for i in x:
        if type(i) == list:
            result += flatten(i)
        else:
            result += [i]
    return result
```

```python
def flat(x):
    tmp = []
    new_data = []
    for i in range(len(x)):
        if type(x[i]) == list:
            tmp.extend(x[i])
        elif type(x[i]) == int:
            tmp.append(x[i])
    for j in range(len(tmp)):
        if type(tmp[j]) == list:
            new_data.extend(tmp[j])
        elif type(tmp[j]) == int:
            new_data.append(tmp[j])
    return new_data
```

```python
def flata(x):
    tmp = []
    for i in range(len(x)):
        if type(x[i]) == list:
            tmp.extend(x[i])
        elif type(x[i]) == int:
            tmp.append(x[i])
    return tmp
```
