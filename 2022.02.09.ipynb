```python
import random
names = "가나다라마바사아자차카타파하"
with open("student_info.csv", "w") as f:
    for i in range(1000):
        name = random.choice(names) + random.choice(names) + random.choice(names)
        height = random.randrange(150, 190)
        weight = random.randrange(40, 90)
        f.write(f'{name},{height},{weight}\n')
```

```python
with open("student_info.csv", "r") as f:
    for line in f:
        (name, height, weight) = line.strip().split(",")
        bmi = int(weight) / int(height)**2
        result = ""
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상"
        else:
            result = "저체중"
        print('\n'.join([
            "이름: {}",
            "몸무게: {}",
            "키: {}",
            "bmi: {}",
            "결과: {}"
        ]).format(name, weight, height, bmi, result))
```

```python
# 홀수만 출력하기

numbers = list(range(1, 10 + 1))

print("# 홀수만 출력하기")
odds = []
for number in numbers:
    if number % 2 != 0:
        odds.append(number)
print(odds)

odds = [ number for number in numbers if number % 2 != 0 ]
print(odds)

print(list(filter(lambda x : x % 2 != 0,numbers)))
```

```python
## 1~10000까지의 숫자중에서
## random하게 3자리의 수를 100개만 추출해서 data라는 리스트를 만들고
## input함수를 사용해서 수치를 입력받아 타입이 다르면 타입에러 ==> "타입에러"
## 선택한 숫자가 data에 없으면 인덱스에러 ==> "인덱스에러"
## 선택한 숫자가 있는 곳의 index 번호를 반환

import random

data_tmp = list(range(101,1000))
data = []
for i in range(100):
    tmp = random.choice(data_tmp)
    data.append(tmp)
```

```python
input_tmp = input("데이터를 입력하세요")
try:
    input_tmp = int(input_tmp)
    ind = data.index(input_tmp)
    print(ind)
except TypeError as e:
    print("틀린 타입 : ", type(e))
    print("틀린 내용 : ", e)
except ValueError as e:
    print(data.index(random.choice(data)))
```
